```python
#definition of the canvas image
from PIL import Image, ImageDraw
width,height=400,800
image= Image.new('RGB',(width,height),"white")
draw=ImageDraw.Draw(image)
step=9
#Height for the character bars
BAR_TOP=10
BAR_BOTTOM=350
#Height for the space bars
SHORTER_BAR_TOP=150
SHORTED_BAR_BOTTOM=250
```

```python
# the high is fixed in the canvas the width of the characters changes
#Definition of the width bars for each char
def get_bar_width(ch):
    ch = ch.lower()  # convert to lower case
    if ch == " ":
        return 1  # space
    elif "a" <= ch <= "z":  # comparison between strings
        position = ord(ch) - ord("a")
        return position + 2
    else:
        return 1 # no-alphabetic and is not space bar
#Encoding of the text in an image
#The draw is not a local call but is a global call on the image created before
def draw_barcode(text):
    x=10
    for ch in text:
        bar_width=get_bar_width(ch)
        if (ch== " "):
            draw.rectangle((x,SHORTER_BAR_TOP,x+bar_width,SHORTED_BAR_BOTTOM),fill="black")
        else :
            draw.rectangle((x,BAR_TOP,x+bar_width,BAR_BOTTOM),fill="black")
        x+=bar_width+step

```

```python
import os

input_text="Abbas 12/Cheddad"
draw_barcode(input_text)
image.show()
image_name="Output.png"
os.makedirs("data/image", exist_ok=True)
image_path = os.path.join("data/image", image_name)
image.save(image_path)
print("saved image as "+image_name+"")
```

```python
# get the dimension of the canvas
#definition of the canvas image
from PIL import Image, ImageDraw
width,height=400,800
image= Image.new('RGB',(width,height),"white")
draw=ImageDraw.Draw(image)
step=9
#Height for the character bars
BAR_TOP=10
BAR_BOTTOM=350
#Height for the space bars
SHORTER_BAR_TOP=150
SHORTED_BAR_BOTTOM=250
black=(0,0,0)
```

```python
def encode_bar(inn_width,png_file,x)->str:
    if(inn_width==1 and png_file.getpixel((x-1,BAR_BOTTOM)[:3])==black):
        return "?"
    elif(inn_width==1 and png_file.getpixel((x-1,SHORTED_BAR_BOTTOM)[:3])==black):
        return " "
    else:
        inn_width-=2
        inn_width=inn_width+ord("a")
        return chr(inn_width)

# scan of the image
def decode_image(png_file):
   str_out=""
   limit_reached=False
   x=11
   inn_width=0
   while not limit_reached:
       if(x>=width):
           limit_reached= True
       else:
           if(png_file.getpixel((x,SHORTER_BAR_TOP))[:3]==black):
               inn_width+=1
               x+=1
           else:
               str_out += encode_bar(inn_width, png_file, x)
               inn_width=0
               x+=step # the char is finished
   return str_out
```

```python
#get the images
import os

folder_path = "data/image"

# List all files in the folder
all_files = os.listdir(folder_path)

# Filter only PNG images
search_image_name="Output.png"
decoded_text=decode_image(Image.open(os.path.join(folder_path,search_image_name)).convert("RGB"))

print("Saved images in folder:", decoded_text)
```

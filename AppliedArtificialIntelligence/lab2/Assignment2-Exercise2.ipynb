```python
import statistics
from typing import List, Tuple, Dict

class StopCriterion:
    def __init__(self,
                 eps_rel: float = 1e-3,
                 eps_abs: float = 1e-2,
                 patience: int = 50,
                 top_k: int = 10,
                 div_factor: float = 1e-4,
                 near_tol: float = 5e-3):
        """
        eps_rel: minimum relative improvement considered significant (e.g., 1e-3 -> 0.1%)
        eps_abs: minimum absolute improvement considered significant
        patience: consecutive generations without improvement to stop
        top_k: how many top solutions to consider when computing diversity/std
        div_factor: relative threshold (std / best) below which diversity is considered low
        near_tol: relative 'closeness' threshold for accepting multiple similar solutions
        """
        self.eps_rel = eps_rel
        self.eps_abs = eps_abs
        self.patience = patience
        self.top_k = max(1, top_k)
        self.div_factor = div_factor
        self.near_tol = near_tol

        # internal state
        self.best_history: List[float] = []
        self.no_improve = 0
        self.last_best = None

    def _top_scores(self, population) -> List[float]:
        """Returns the fitness values (float) of the top_k elements of the population (assumed sorted descending)."""
        top = population[:self.top_k]
        return [float(getattr(ind, "hit_walls", 0.0)) for ind in top]

    def _top_genome_uniqueness(self, population) -> float:
        """
        Simple diversity measure: fraction of unique genomes among the top_k.
        Assumes each individual has .moves (a list). Convert to tuple for comparison.
        Returns a value in [0,1].
        """
        top = population[:self.top_k]
        genomes = []
        for ind in top:
            try:
                g = tuple(ind.moves)
            except Exception:
                # fallback: string representation
                g = str(ind.moves)
            genomes.append(g)
        if not genomes:
            return 0.0
        unique = len(set(genomes))
        return unique / len(genomes)

    def update(self, population: List[object], generation: int) -> Tuple[bool, str, Dict]:
        """
        population: list of Individual objects sorted descending by fitness (best first)
        generation: current generation number (integer)
        Returns (should_stop, reason, info)
        info contains: best, prev_best, improvement, rel_improvement, top_std, top_mean, top_uniqueness, no_improve
        """
        if not population:
            return True, "empty_population", {"best": None}

        # extract top_k scores (population assumed already sorted descending)
        top_scores = self._top_scores(population)
        best = top_scores[0]

        # update history
        prev_best = self.last_best
        self.best_history.append(best)
        self.last_best = best

        # top-k statistics
        top_mean = statistics.mean(top_scores) if top_scores else float(best)
        top_std = statistics.pstdev(top_scores) if len(top_scores) > 1 else 0.0

        # diversity among top-k (0..1)
        top_uniqueness = self._top_genome_uniqueness(population)

        # absolute and relative improvement compared to previous best known
        improvement = 0.0
        rel_improvement = 0.0
        if prev_best is not None:
            improvement = best - prev_best
            # if prev_best is close to zero, use denominator 1.0 to avoid division by zero
            denom = abs(prev_best) if abs(prev_best) > 1e-12 else 1.0
            rel_improvement = improvement / denom

        # check significant improvement
        improved = (rel_improvement >= self.eps_rel) or (abs(improvement) >= self.eps_abs)

        if improved:
            self.no_improve = 0
        else:
            self.no_improve += 1

        # criterion 1: no improvement for 'patience' generations
        if self.no_improve >= self.patience:
            reason = f"no_improvement (patience={self.patience})"
            info = {
                "best": best,
                "prev_best": prev_best,
                "improvement": improvement,
                "rel_improvement": rel_improvement,
                "top_std": top_std,
                "top_mean": top_mean,
                "top_uniqueness": top_uniqueness,
                "no_improve": self.no_improve
            }
            return True, reason, info

        # criterion 2: top-k too similar (low diversity)
        # consider both relative standard deviation and fraction of unique genomes
        rel_std = (top_std / (abs(top_mean) + 1e-9))
        if rel_std < self.div_factor and top_uniqueness < (1.0 - self.near_tol):
            # if both std and uniqueness are low -> stop
            reason = f"low_diversity (rel_std={rel_std:.6g}, uniqueness={top_uniqueness:.3f})"
            info = {
                "best": best,
                "prev_best": prev_best,
                "improvement": improvement,
                "rel_improvement": rel_improvement,
                "top_std": top_std,
                "top_mean": top_mean,
                "rel_std": rel_std,
                "top_uniqueness": top_uniqueness,
                "no_improve": self.no_improve
            }
            return True, reason, info

        # criterion 3: convergence (more conservative): if top solutions fall within near_tol
        # consider stopping acceptable if relative difference between top_k and best is very small
        if len(top_scores) >= 2:
            # fraction of top within near_tol relative to best
            count_near = sum(1 for s in top_scores if abs(best - s) / (abs(best) + 1e-9) <= self.near_tol)
            frac_near = count_near / len(top_scores)
            if frac_near >= 0.9 and rel_std < (self.div_factor * 10):
                reason = f"tops_converged (frac_near={frac_near:.2f})"
                info = {
                    "best": best,
                    "prev_best": prev_best,
                    "improvement": improvement,
                    "rel_improvement": rel_improvement,
                    "top_std": top_std,
                    "top_mean": top_mean,
                    "top_uniqueness": top_uniqueness,
                    "frac_near": frac_near,
                    "no_improve": self.no_improve
                }
                return True, reason, info

        # otherwise, do not stop
        info = {
            "best": best,
            "prev_best": prev_best,
            "improvement": improvement,
            "rel_improvement": rel_improvement,
            "top_std": top_std,
            "top_mean": top_mean,
            "top_uniqueness": top_uniqueness,
            "no_improve": self.no_improve
        }
        return False, "continue", info
```

```python
import copy
import math

"""
    Revive in input a maze matrix and return the path to the exit
"""
import random
DIRS={"U": (-1,0),"D": (1,0),"L": (0,-1),"R": (0,1)}
POPUL_SIZE = 500
MUTATION_RATE = 0.1

stopper = StopCriterion(
    eps_rel=1e-3,     # 0.1% relative improvement
    eps_abs=0.05,     # absolute improvement threshold
    patience=5,      # generations without improvement before stopping
    top_k=10,         # number of top solutions to consider for diversity/std
    div_factor=1e-4,  # diversity threshold
    near_tol=5e-3     # 0.5% closeness tolerance
)

start = Coord(0, 0)
target = Coord(12,36)
maze=maze_matrix

def in_bounds(p:Coord):
    #len(maze) is the number of rows
    #len(maze[0]) is the len ( coloumn ) of the  first row
    return 0<=p.row<len(maze) and 0<=p.col<len(maze[0])

def dist(a:Coord,b:Coord)  :
    return abs(a.row-b.row)+abs(a.col-b.col)


def get_neighbours(p:Coord):
    neighbours=[]
    for dir, (dr, dc) in DIRS.items():
        r, c = p.row + dr, p.col + dc
        if in_bounds(Coord(r, c)):
            neighbours.append(Coord(r, c))
    return neighbours


def path_to_target_simple(start: Coord, end: Coord) -> list[str]:
    """
    Generate a simple sequence of moves from start to end.
    """
    moves = []
    current = Coord(start.row, start.col)

    while current != end:
        dr = end.row - current.row
        dc = end.col - current.col

        options = []
        if dr != 0:
            options.append("D" if dr > 0 else "U")
        if dc != 0:
            options.append("R" if dc > 0 else "L")
        if options:
            step = random.choice(options)
            current = Coord(current.row + DIRS[step][0], current.col + DIRS[step][1])
            moves.append(step)
    return moves



def random_valid_coord(maze=maze) -> Coord:
    while True:
        row = random.randint(0, len(maze) - 1)
        col = random.randint(0, len(maze[0]) - 1)
        if maze[row][col] == 1:
            return Coord(row, col)

def get_max_moves(n_gen:int=1):
    # base = Manhattan distance * factor
    base = dist(start, target)
    #slow increase of the valid moves
    extra = (n_gen // 50) * 5
    max_len = max(5, base + 20 + extra)
    return max_len


class Individual:
    """
    Rappresent the individual and its genome, for each one the path should be evaluated only once, is always the same, the moves are stored in the move attribute and they're not changing
    """
    def __init__(self,moves:list[str], n_gen:int,reached:bool=False, last:Coord=start):
        self.reached=reached
        self.moves=moves
        self.num_cycles=0
        self.move_out_of_bounds=0
        self.hit_walls=len(maze[0])*len(maze)
        self.last=last
        self.path=[]
        self.fitness=self.fitness_of_moves(n_gen)



    @classmethod
    def create_seq_moves_random(cls, n_gen:int)-> list[DIRS.keys()]:
        """
        Create a random sequence of moves (genome).
        Length = Manhattan distance * factor
        used for initialization of the population
        """
        #base_len = dist(start, target)
        base_len = get_max_moves(n_gen)
        length = random.randint(base_len-5, int(base_len * 1))  # allow some flexibility
        genome = [random.choice(list(DIRS.keys())) for _ in range(length)]
        return genome

    @classmethod
    def create_seq_moves_sequence(cls, n_gen:int)-> list[DIRS.keys()]:
        """
        Create a sequence of moves (genome) directed to a destination
        Length = Manhattan distance * factor
        used for initialization of the population
        """
        #base_len = dist(start, target)
        base_len = get_max_moves(n_gen)
        length = random.randint(base_len-5, int(base_len * 1))  # allow some flexibility
        x_coord=random.randint(0,len(maze)-1)
        y_coord=random.randint(0,len(maze[0])-1)
        random_point=Coord(x_coord,y_coord)
        genome1=path_to_target_simple(start,random_point)
        genome2=path_to_target_simple(random_point,target)
        genome=genome1+genome2
        genome = genome[:length]
        return genome

    def mutation_genes(self):
        # return a random move
        return random.choice(list(DIRS.keys()))



    def mate(self,par2:"Individual", n_gen:int=0) -> "Individual":
        # generate the mutation with only a single point of join for the 2 parents
        max_len=get_max_moves(n_gen)
        len_min=min(len(self.moves), max_len,len(par2.moves))
        base_len=random.randint(len_min-5, len_min)
        child_chromosome = self.moves[:base_len] + par2.moves[base_len:len_min]
        # apply of the mutation
        child_chromosome_new = []
        act_coord = Coord(start.row, start.col)
        mutation_rate = MUTATION_RATE
        if n_gen > 1000:
            mutation_rate = min(0.3, math.log10(n_gen/100) * 1.5)
        i = 0
        while i < len(child_chromosome):
            mv = child_chromosome[i]
            dr, dc = DIRS[mv]
            next_coord = Coord(act_coord.row + dr, act_coord.col + dc)
            if in_bounds(next_coord) and random.random() < mutation_rate:
                # choose a sequence of moves with length in [2, 4]
                seq_len = random.randint(2, 4)
                for _ in range(seq_len):
                    mut_mv = self.mutation_genes()
                    child_chromosome_new.append(mut_mv)
                    act_coord = Coord(act_coord.row + DIRS[mut_mv][0], act_coord.col + DIRS[mut_mv][1])
                i += seq_len  # salta avanti di seq_len
            else:
                child_chromosome_new.append(mv)
                act_coord = Coord(act_coord.row + DIRS[mv][0], act_coord.col + DIRS[mv][1])
                i += 1
        return Individual(moves=child_chromosome_new,n_gen=n_gen)


    def simulate_path(self, max_steps=None)->List[Coord]:
        if (self.path!=[]) :
            return self.path
        self.reached = False
        path = [start]
        visited = {start:0}  # coord -> index
        max_steps = max_steps or len(maze) * len(maze[0])
        local_hit_walls=0
        local_move_out_of_bounds=0
        local_num_cycles=0
        i = 0
        while i < min(max_steps,len(self.moves)):
            if not path:
                path = [start]
            if self.reached :
                break
            mv = self.moves[i]
            dr, dc = DIRS[mv]
            act_coord = Coord(path[-1].row + dr, path[-1].col + dc)
            if act_coord == target :
                path.append(target)
                self.reached = True
                break
            # out of bounds
            if not in_bounds(act_coord):
                local_move_out_of_bounds += 1
                i+=1
                continue
            # hit a wall
            if  maze[act_coord.row][act_coord.col] == 0:
                local_hit_walls+=1
            # found a loop
            if act_coord in visited:
                #correct the loop from the start
                loop_start = visited[act_coord]
                # remove the eventuals nodes in the middle of the loop
                for old_idx in range(loop_start, len(path)):
                    visited.pop(path[old_idx], None)
                #truncation of the path
                path = path[:loop_start]
                local_num_cycles += 1
            else:
                path.append(act_coord)
                visited[act_coord] = len(path) - 1
            i += 1
        if not path:
                path = [start]



        if(self.hit_walls>local_hit_walls and self.path==[]):
            self.last = path[-1]
            self.hit_walls=local_hit_walls
            self.move_out_of_bounds=local_move_out_of_bounds
            self.num_cycles=local_num_cycles
            self.path=path
        return path


    def fitness_of_moves(self, n_gen) -> float:
        """
        evaluation of the fitness
        """

        max_moves = get_max_moves(n_gen)
        gen_idx = max(1, n_gen)
        bm=BestMetrics.get_instance()
        #value that increase with the generation
        gen_factor = min(1.0, math.log10(n_gen + 1) / 10.0)
        # rebuild the path if it's not already done' from the sequence of moves
        if(self.path==[]):
            path = self.simulate_path(max_steps=max_moves)
        else:
            path=self.path
        # get the actual distance from the target
        dist = abs(self.last.row - target.row) + abs(self.last.col - target.col)
        l = max(1, len(self.moves))
        # update the best metrics if the current solution is better than the previous one
        if bm.get_best_hit_walls()>self.hit_walls and self.reached== True:
            bm.set_individual(self,n_gen)


        # ----- TERM: proximity (the more is closer the better) -----
        # the closer score is just if the distance is less than 5 so that the chromosomes remain in that distance range
        prox = 1.0 / (1.0 + dist)
        prox_term=0
        if(dist<5):
            prox_term = 2000.0 * prox

        # ----- TERM: improvement if move from the start -----
        # the more is closer the better, and the chromosome doesn't assume the easy path to not get the penality terms
        dist_start = abs(start.row - target.row) + abs(start.col - target.col)
        if dist_start <= 0:
            improvement_frac = 1.0 if dist == 0 else 0.0
        else:
            improvement = max(0, dist_start - dist)
            improvement_frac = improvement / float(dist_start)
        improvement_term = 50 * improvement_frac

        # ----- TERM: improvement if the path visits nodes -----
        unique_visited = len(set(path))
        cap_nodes = min(2 * dist_start, max_moves * 0.8)
        effective_nodes = min(unique_visited, cap_nodes)

        k = max(4.0, dist_start / 2.0)  # scala
        explore_gain_frac = 1.0 - math.exp(- effective_nodes / k)

        usefulness_scale = 1.0 / (1.0 + dist)
        length_scale = (l / float(max_moves)) if max_moves > 0 else 1.0

        explore_term = 80.0 * explore_gain_frac * length_scale * (1.0 + (1.0 - gen_factor)) * (0.5 + 0.5 * usefulness_scale)

        # ----- TERM: penality of invalid moves-----
        invalid_contrib = (
            1*float(self.num_cycles)
            + 5.0 * float(self.hit_walls) * 1/len(path)
            + 10 * (self.move_out_of_bounds / (math.log1p(l) + 1e-9))
        )
        inv_norm = math.tanh(invalid_contrib / 8.0)
        invalid_term = 10*inv_norm * gen_factor

        # ----- TERM: penality of a short path that not reach the target -----
        short_penalty = 0.0
        fill_penalty = 0.0
        if not self.reached:
            short_threshold = dist + 1
            if l < short_threshold:
                short_penalty = 50.0 * (1.0 - (l / float(short_threshold))) * gen_factor
            fill_frac = max(0.0, (max_moves - l) / float(max_moves)) if max_moves > 0 else 0.0
            fill_penalty = 3.0 * fill_frac * gen_factor

        # ----- TERM: bonus length if the invalidity is low -----
        length_bonus = 0.0
        if invalid_contrib < 3.0:
          length_bonus = 400.0 * (math.log1p(l) / (math.log1p(max_moves) + 1e-9)) * explore_gain_frac

        # activate therms if the destination is reached
        active=1
        if(self.reached==True):
            active=0
        score = (
            prox_term*active
            + improvement_term
            + explore_term
            + length_bonus*active
            - invalid_term
            - short_penalty
            - fill_penalty
        )


        # ----- TERM: target reached -----
        if self.reached and path and path[-1] == target:
            if self.reached and path[-1] == target:
                inv_penalty = 1.0 / (1.0 + gen_factor*self.hit_walls)
                score = score * 1000.0 + 2000.0 * inv_penalty


        self.fitness = float(score)
        return float(score)



class BestMetrics:
    _instance = None   # class variable to store the single instance of the best class

    def __init__(self):
        self.individual=None
        self.generation=0


    @classmethod
    def get_instance(cls):
        if cls._instance is None:
            cls._instance = BestMetrics()
        return cls._instance

    def set_individual_void(self):
        self.individual=Individual(moves=[],last=start,n_gen=0)
        self.individual.hit_walls=len(maze[0])*len(maze)

    def set_individual(self, individual_inp: "Individual",n_gen:int):
        if individual_inp.path==[] :
            return
        cond_new_better = False
        if (self.individual is None or  individual_inp.hit_walls<self.individual.hit_walls
            and individual_inp.path[0]==start and individual_inp.path[-1]==target) :
           cond_new_better = (
            self.individual is None or
            (individual_inp.hit_walls < self.individual.hit_walls
             and individual_inp.path
             and individual_inp.path[0] == start
             and individual_inp.path[-1] == target)
        )
        if cond_new_better:
            # Deep copy for doesn't share the reference of the object
            self.individual = copy.deepcopy(individual_inp)
            self.individual.reached = True
            self.generation = n_gen
        else :
            return

    def get_individual(self) -> "Individual":
        return copy.deepcopy(self.individual)

    def get_best_hit_walls(self) -> int:
        return copy.deepcopy(self.individual).hit_walls if self.individual else float("inf")




def main():
    population = []
    bm=BestMetrics.get_instance()
    bm.set_individual_void()

    for _ in range(POPUL_SIZE):
        moves = Individual.create_seq_moves_random(0)  # genome from start → mid → target
        ind = Individual(moves=moves, last=start,n_gen=0)
        population.append(ind)
    should_stop, reason, info = stopper.update(population,0)
    prec_score: int = population[0].fitness
    same_value=0;
    generation = 1
    found = False

    while not found:
        # Sort by the best fitness score in descending order of the fitness value
        population = sorted(population, key=lambda x: x.fitness, reverse=True)

        if prec_score < population[0].fitness:
            same_value=0
            print(
                f"Gen: {generation}\tSolution: {' -> '.join(str(c) for c in population[0].path)}\t"
                f"Fitness Score: {population[0].fitness}        Hit walls for : {population[0].hit_walls}"
            )
            print("wall in the best metrics :",bm.get_best_hit_walls())
        else :
            same_value+=1
        prec_score = population[0].fitness
        generation += 1
        # check 1 of the stopping criterion
        if( bm.get_individual().hit_walls==0):
            print(
                f"Gen: {bm.generation}\tSolution: {' -> '.join(str(c) for c in bm.individual.path)}\t"
                f"Fitness Score: { ""}        Hit walls for : {bm.individual.hit_walls}"
            )
            found=True
        # check 2 of the stopping criterion
        if should_stop and population[0].reached :
            print(f"Gen {generation} | best={info['best']:.3f} | rel_imp={info['rel_improvement']:.6f} "
                  f"| top_std={info['top_std']:.3f} | no_impr={info['no_improve']} | reason={reason}")
            print("Stopping criterion met:", reason)
            print(
                f"Gen: {bm.generation}\tSolution: {' -> '.join(str(c) for c in bm.individual.path)}\t"
                f"Hit walls for : {bm.individual.hit_walls}"
            )
            break



        # create a new generation of individuals by crossover and mutation
        new_generation = []
        s = int((10 * POPUL_SIZE) / 100)
        new_generation.extend(population[:s])
        s = int((90 * POPUL_SIZE) / 100)
        parent_pool = population[:POPUL_SIZE // 5]
        for _ in range(s):
            parent1 = random.choice(parent_pool)
            parent2 = random.choice(parent_pool)
            child = parent1.mate(parent2,generation)
            new_generation.append(child)

        #reinjection of individuals
        if generation % 100 == 0:
            print(f"actual n of generation:  {(generation)}")
            s = int((90 * POPUL_SIZE) / 100)
            new_generation=population[:s]
            s = POPUL_SIZE-s-1
            for i in range(s):
                if i%2==0:
                    chromosomes=Individual.create_seq_moves_random(n_gen=generation)
                else :
                    chromosomes=Individual.create_seq_moves_sequence(n_gen=generation)

                child = Individual(moves=chromosomes,n_gen=generation)
                new_generation.append(child)
            best_ind=bm.get_individual()
            new_generation.append(best_ind)
            population=new_generation
            for ind in new_generation:
                ind.fitness = ind.fitness_of_moves(n_gen=generation)
            prec_score = 0  #reset prec best value
            if(population[0].last==target): should_stop, reason, info = stopper.update(population,0)
        population = new_generation

if ( __name__=="__main__" ):
    main()
```
